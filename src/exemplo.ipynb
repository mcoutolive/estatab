```python
from estatab import HipoteseSimplificada

# Cenário: tempo médio de processamento (ms)
hip = HipoteseSimplificada(
    nome_metrica="tempo_ms",
    tipo_metrica="numerica",          # ou "media"
    tamanho_amostral_calculado=1000,
    intencao_teste="detectar_reducao",# H1: tratamento < controle
    metodologia="frequentista",
    tipo_correcao="Bonferroni",
)

# Controle (A)
hip.adiciona_variante(
    nome_variante="A_controle",
    numero_amostras=1000,
    valor_media=210.0,
    desvio_padrao=55.0,
    variante_controle=True
)

# Tratamento (B)
hip.adiciona_variante(
    nome_variante="B_trat",
    numero_amostras=1000,
    valor_media=195.0,
    desvio_padrao=50.0,
    variante_controle=False
)

resultado = hip.executa_teste_hipotese()
resultado

```
